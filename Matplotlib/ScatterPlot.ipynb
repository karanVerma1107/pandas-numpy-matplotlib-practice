```python
# Import necessary libraries for plotting and numerical operations
import matplotlib.pyplot as plt  # For plotting graphs
import numpy as np               # For numerical calculations and array handling

# Use a clean style for plots with gridlines 
plt.style.use('ggplot')

# Create 30 evenly spaced numbers between 0 and 10 (x values)
x = np.linspace(0, 10, 30)

# Calculate the sine of each x value (y values)
y = np.sin(x)

# Scatter plot using plt.plot:
# 'o' means marker is a circle
# color='black' sets the color of the markers
plt.plot(x, y, 'o', color='black')
plt.show()

# Create a random number generator for reproducible random numbers
rng = np.random.RandomState(0)
# Loop over a list of different marker styles
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    # Generate 5 random x-coordinates using rng.rand(5)
    # rng.rand(5) produces 5 random numbers between 0 and 1
    x_values = rng.rand(5)
    
    # Generate 5 random y-coordinates using rng.rand(5)
    y_values = rng.rand(5)
    
    # Plot the 5 points on the graph with the current marker style
    # 'marker' sets the shape of the points (like circles, dots, crosses, etc.)
    # label sets the legend entry to describe which marker is used
    plt.plot(x_values, y_values, marker, label=f"marker='{marker}'")

# After this loop, each iteration adds a scatter plot of 5 points with a different marker
# The result is a combined plot showing multiple marker types with random points

# Show a legend that describes each marker style
plt.legend(numpoints=1)  # Show one marker per label in the legend

# Set x-axis limits so all points are nicely visible
plt.xlim(0, 1.8)
plt.show()



# Plot line with markers combined:
# '-' connects points with a line
# 'o' places circle markers on the line
# 'k' stands for black color
plt.plot(x, y, '-ok')
plt.show()

# More customized plot with different line and marker properties:
plt.plot(x, y, '-p',            # Line with pentagon markers ('p')
         color='gray',          # Gray color for line and marker edges
         markersize=15,         # Marker size bigger (15 points)
         linewidth=4,           # Thickness of the connecting line
         markerfacecolor='white',   # Inside of marker is white (hollow look)
         markeredgecolor='gray',    # Marker edges are gray
         markeredgewidth=2)          # Thickness of the marker edges

# Set y-axis limits for better viewing range
plt.ylim(-1.2, 1.2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a random number generator with fixed seed for reproducibility
rng = np.random.RandomState(0)

# 2. Generate 100 random points for x-coordinates from a normal distribution (mean=0, std=1)
x = rng.randn(100)

# 3. Generate 100 random points for y-coordinates from the same distribution
y = rng.randn(100)

# 4. Generate 100 random values between 0 and 1 to represent colors for each point
colors = rng.rand(100)

# 5. Generate 100 random sizes, scaled up to 1000 pixels for each point
sizes = 1000 * rng.rand(100)

# 6. Create scatter plot with:
#    - x and y coordinates
#    - colors mapped from the 'colors' array, with colormap 'viridis'
#    - sizes specified by 'sizes' array
#    - alpha (transparency) set to 0.3 to better see overlapping points
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')

# Set x-axis limits to show range from -3 to 3
plt.xlim(-3, 3)

# Set y-axis limits to show range from -3 to 3
plt.ylim(-3, 3)

# 7. Add a colorbar to show the mapping between colors and the numerical color values
plt.colorbar()

plt.show()
```


When you have a small dataset, using `plt.plot` or `plt.scatter` doesnt make much difference. But with very large datasets (thousands of points or more), `plt.plot` is faster because it treats all points the same way and draws them quickly.

`plt.scatter` is slower for big data because it can give each point its own size and color, which takes more work to draw.

So, for large datasets, its better to use `plt.plot` if you dont need different colors or sizes for each point.
