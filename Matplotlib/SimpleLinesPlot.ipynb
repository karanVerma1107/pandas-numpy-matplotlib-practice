```python
# 1. This enables inline plotting in Jupyter Notebook (you can skip it in normal .py files)
# %matplotlib inline  # Uncomment only if you're using Jupyter Notebook

# 2. Import matplotlib for plotting and numpy for numerical calculations
import matplotlib.pyplot as plt
import numpy as np

# 3. Optional: Use a nicer plot style with grid and clean background
plt.style.use('ggplot')          # Clean with grid and red tones
# plt.style.use('classic')       # Original Matplotlib style (basic)
# plt.style.use('bmh')           # "Bayesian Methods for Hackers" style (nice and readable)
# plt.style.use('fivethirtyeight')  # Inspired by 538.com plots
# plt.style.use('dark_background')  # Dark theme (great for presentations)

# -----------------------------
# Step 1: Create a figure and an axes manually
# -----------------------------
fig = plt.figure()          # 4. Create a blank figure (like a canvas)
ax = plt.axes()             # 5. Create an axes (the box where the graph will appear)

# 6. Create X values from 0 to 10 (1000 points in between)
x = np.linspace(0, 10, 30) # This generates 1000 evenly spaced points between 0 and 10
# 

# s just defining the x-coordinates from 0 to 10 — it’s not the origin or both axes, just a set of points along the x-axis.

# 7. Plot y = sin(x) on the axes
ax.plot(x, np.sin(x))       # This draws the line for sin(x)

# -----------------------------
# Step 2: Do the same using simpler shortcut — no need to create fig and ax
# -----------------------------
# 8. This automatically creates a figure and axes in the background
plt.plot(x, np.sin(x))      # Still shows sin(x)

# -----------------------------
# Step 3: Plot multiple lines on the same graph
# -----------------------------
# 9. First plot: y = sin(x)
plt.plot(x, np.sin(x))

# 10. Second plot: y = cos(x) — this adds another line on the same figure
plt.plot(x, np.cos(x))


# 11. Show the final plot with both lines (not required in Jupyter, but good practice in scripts)
plt.show()
```



### What exactly is this **1000** in `np.linspace(0, 10, 1000)`?

* **1000** is the number of points or samples you want between 0 and 10.
* Imagine you want to draw a curve on your computer. The computer doesn’t draw a continuous line by itself — it draws **many tiny dots** and connects them.
* **1000 means**: "Give me 1000 dots spread evenly from 0 to 10 on the x-axis."
* The more dots you have, the smoother and more accurate your line looks.

---

### What about real paper and pen — how do these numbers relate?

When you draw a graph on paper:

* You usually **mark some points** on the x-axis (like 0, 1, 2, ..., 10).
* Then you calculate the y-values for those points (like y = sin(x)).
* You **connect those points with a smooth curve** using your hand.
* You don't actually plot thousands of points — your brain **smooths the curve** automatically!

---

### So how does this relate?

* The **“1000”** in code is like saying: "Draw a very detailed, very smooth curve by plotting lots of points."
* On paper, you don’t plot 1000 points, but your hand and eye **approximate the curve smoothly** between the points you mark.
* On a computer, the program needs many points (like 1000) to create that smoothness **because it draws dots and connects them.**

---

### Summary:

* **1000** = number of points (dots) computer uses to draw the curve.
* On paper, you mark fewer points but your hand smooths the curve.
* More points on computer = smoother plot; fewer points = jagged or rough lines.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create X values from 0 to 10 (1000 points)
x = np.linspace(0, 10, 1000)

# ----------------------------------------
# Plotting lines with different color specifications
# ----------------------------------------

# Using full color name as string
plt.plot(x, np.sin(x - 0), color='blue')       

# Using single character color code (r=red, g=green, b=blue, etc.)
plt.plot(x, np.sin(x - 1), color='g')           

# Using grayscale as string between 0 (black) and 1 (white)
plt.plot(x, np.sin(x - 2), color='0.75')        

# Using Hex color code (RGB hex string, e.g. '#RRGGBB')
plt.plot(x, np.sin(x - 3), color='#FFDD44')     

# Using RGB tuple with values between 0 and 1 for Red, Green, Blue channels
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) 

# Using named HTML colors (supported by Matplotlib)
plt.plot(x, np.sin(x - 5), color='chartreuse')  

# ----------------------------------------
# Plotting lines with different line styles
# ----------------------------------------
plt.figure()
# Solid line (default)
plt.plot(x, x + 0, linestyle='solid')   

# Dashed line
plt.plot(x, x + 1, linestyle='dashed')  

# Dash-dot line
plt.plot(x, x + 2, linestyle='dashdot') 

# Dotted line
plt.plot(x, x + 3, linestyle='dotted') 

# ----------------------------------------
# Shorter line style codes (equivalent to above)
# ----------------------------------------
plt.figure()
plt.plot(x, x + 4, linestyle='-')   # solid
plt.plot(x, x + 5, linestyle='--')  # dashed
plt.plot(x, x + 6, linestyle='-.')  # dashdot
plt.plot(x, x + 7, linestyle=':')   # dotted

# ----------------------------------------
# Combining color and linestyle as a single positional argument
# ----------------------------------------

# This string contains line style and color codes combined
# '-g' means solid line ( - ) and green (g)
plt.plot(x, x + 0, '-g')  

# '--c' means dashed line and cyan color
plt.plot(x, x + 1, '--c') 

# '-.k' means dash-dot line and black color
plt.plot(x, x + 2, '-.k') 

# ':r' means dotted line and red color
plt.plot(x, x + 3, ':r')  

plt.show()

# ----------------------------------------
# Summary / How parameters work:
# ----------------------------------------
# plt.plot() function draws lines on a graph.
# It takes multiple arguments:
#   - The first is the X data points.
#   - The second is the Y data points.
#   - Additional keyword arguments like 'color' and 'linestyle' control how the line looks.
#
# Color can be passed as:
#   - A named string ('blue')
#   - A short code ('g' for green)
#   - A grayscale value ('0.75')
#   - A hex code ('#FFDD44')
#   - An RGB tuple ((1.0, 0.2, 0.3))
#
# Line style controls the pattern of the line, e.g. solid, dashed, dotted.
#
# Color and linestyle can also be combined as one shorthand string like '-g' for green solid line.
#
# This flexibility lets you easily customize plots for clarity, style, and presentation.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

# Basic plot of sin(x)
plt.plot(x, np.sin(x))

# Set X axis limits from -1 to 11 (start a little before 0 and end a little after 10)
plt.xlim(-1, 11)

# Set Y axis limits from -1.5 to 1.5 (wider than sin range to give space)
plt.ylim(-1.5, 1.5)

plt.title("Set xlim and ylim with normal order")
plt.show()
```

```python
plt.plot(x, np.sin(x))

# Reverse X axis limits: start from 10 and go to 0 (flips the x-axis)
plt.xlim(10, 0)

# Reverse Y axis limits: start from 1.2 and go to -1.2 (flips the y-axis)
plt.ylim(1.2, -1.2)

plt.title("Reversed axes using xlim and ylim")
plt.show()
```

What happens here?

Axis direction is flipped by reversing the order of the limits.

Use this if you want to invert the axis (e.g., show time backwards or invert coordinates).



```python
plt.plot(x, np.sin(x))

# Set all axis limits at once: [xmin, xmax, ymin, ymax]
plt.axis([-1, 11, -1.5, 1.5])

plt.title("Set axes limits using plt.axis()")
plt.show()
```

What happens here?

'tight' tells matplotlib to shrink the axes bounds to fit the data exactly.

Use when you want to remove unnecessary whitespace around your plot.

```python
plt.plot(x, np.sin(x))

# Automatically tighten the axis limits to fit the data tightly
plt.axis('tight')

plt.title("Auto tighten axis limits")
plt.show()
```

What happens here?

'equal' forces the aspect ratio so one unit on x equals one unit on y visually.

Useful when plotting shapes where proportions matter, like circles or squares.

```python
plt.plot(x, np.sin(x))

# Make the x and y axis have the same scale (equal units)
plt.axis('equal')

plt.title("Equal aspect ratio (1 unit x = 1 unit y)")
plt.show()
```

Summary (in simple words):

plt.xlim(a, b) and plt.ylim(c, d) let you control the visible range of the x-axis and y-axis.

You can zoom in/out or flip axes by adjusting these limits.

plt.axis([xmin, xmax, ymin, ymax]) sets all axis limits together.

plt.axis('tight') shrinks the view to fit your data perfectly.

plt.axis('equal') ensures both axes use the same scale for accurate shapes.

# Labeling is below

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)  # Generate 1000 points from 0 to 10

# Plot sin(x) curve
plt.plot(x, np.sin(x))

# Add a title to the plot
plt.title("A Sine Curve")  # This sets the main title on top of the plot

# Label the x-axis
plt.xlabel("x")            # This labels the horizontal axis as "x"

# Label the y-axis
plt.ylabel("sin(x)")       # This labels the vertical axis as "sin(x)"

plt.show()
```

What happens here?

The label parameter in plt.plot() assigns a name to each line.

plt.legend() reads these labels and shows a legend box matching each line's color and style.

Legend helps to understand which line represents what data.

```python
# Plot sin(x) with green solid line and label it "sin(x)"
plt.plot(x, np.sin(x), '-g', label='sin(x)')

# Plot cos(x) with blue dotted line and label it "cos(x)"
plt.plot(x, np.cos(x), ':b', label='cos(x)')

# Set equal scaling on x and y axes (optional, for better shape view)
plt.axis('equal')

# Show the legend (uses the labels from the plot function)
plt.legend()

plt.show()
```

What happens here?

fig, ax = plt.subplots() creates a figure and axes to control plots more flexibly.

ax.plot() plots on the given axes object.

ax.set() can set many properties like limits, labels, and title at once, instead of multiple calls.

This style is preferred for complex plots or multiple subplots.

```python
fig, ax = plt.subplots()  # Create a figure and axes object

ax.plot(x, np.sin(x))      # Plot sin(x) on the axes

# Set multiple properties at once using ax.set()
ax.set(
    xlim=(0, 10),          # Set x-axis limits from 0 to 10
    ylim=(-2, 2),          # Set y-axis limits from -2 to 2
    xlabel='x',            # Label x-axis
    ylabel='sin(x)',       # Label y-axis
    title='A Simple Plot'  # Title of the plot
)

plt.show()
```

Summary:

plt.title(), plt.xlabel(), plt.ylabel() add title and axis labels in the simple pyplot interface.

When plotting multiple lines, add labels with label='name' inside plt.plot().

Call plt.legend() to display these labels as a legend.

The object-oriented approach uses axes objects (ax) and methods like ax.set() to do the same but is more flexible.

Use labeling to make plots understandable and to explain what data each axis and line represents.
