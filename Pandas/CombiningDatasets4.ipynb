suppose we have two datasets A & B, so basically there are four joins that are inner join, outer join, left join and right join.

suppose we have to find all those rows or values  that are common to both of them which is also known as key colomn we can do it with help of Inner join

if we have to find the non-common rows or by intersection of two Datasets then that is Outer join

left join is something when we take intersection of a dataset A to that of dataset B along with the common values or rows present in both.

right join is when we take intersection of A nd B datasets along with common values present in A and non-common of B

```python
import pandas as pd

# üéì First DataFrame: Names and Math marks of 7 students
df_math = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Fay', 'George'],
    'Math_Marks': [85, 78, 92, 88, 76, 90, 81]
})

# üî¨ Second DataFrame: Names and Science marks of 7 students (some same, some different)
df_science = pd.DataFrame({
    'Name': ['Charlie', 'David', 'Eva', 'Fay', 'George', 'Hannah', 'Ian'],
    'Science_Marks': [91, 82, 87, 89, 90, 77, 85]
})

# üîó INNER JOIN: Only names present in both datasets
inner_merged = pd.merge(left=df_math, right=df_science, on='Name', how='inner')
print("üîπ INNER JOIN:")
print(inner_merged)

# üîó OUTER JOIN: All names from both datasets; unmatched values become NaN
outer_merged = pd.merge(left=df_math, right=df_science, on='Name', how='outer')
print("\nüîπ OUTER JOIN:")
print(outer_merged)

# üîó LEFT JOIN: All names from df_math; add matching science marks if available
left_merged = pd.merge(left=df_math, right=df_science, on='Name', how='left')
print("\nüîπ LEFT JOIN:")
print(left_merged)

# üîó RIGHT JOIN: All names from df_science; add matching math marks if available
right_merged = pd.merge(left=df_math, right=df_science, on='Name', how='right')
print("\nüîπ RIGHT JOIN:")
print(right_merged)



# üí° If the column names were different, like 'StudentName' in df_math and 'Name' in df_science:
# You would use: pd.merge(left=df_math, right=df_science, left_on='StudentName', right_on='Name', how='inner')
```

join() in pandas
join() is a method of DataFrame, used to join based on the index by default.

It is typically used when one or both DataFrames have meaningful indexes.

Syntax: df1.join(df2) ‚Äî joins df2 to df1 using their indexes.



| Feature     | `merge()`                                | `join()`                                |
| ----------- | ---------------------------------------- | --------------------------------------- |
| Based on    | One or more columns                      | Index (by default)                      |
| Flexibility | Very flexible (can specify any key)      | Less flexible (mainly index-based)      |
| Typical Use | When joining on columns (like SQL JOINs) | When index is the key to join on        |
| Called on   | `pd.merge(df1, df2, ...)` (standalone)   | `df1.join(df2)` (called on a DataFrame) |



 Quick Tip:
Use merge() when joining on columns.

Use join() when joining on index, or when chaining operations on a DataFrame.

```python
import pandas as pd

# üìò DataFrame 1: Student names as index, and Math marks
df_math = pd.DataFrame({
    'Math_Marks': [85, 78, 92, 56]
}, index=['Alice', 'Bob', 'Kevin','Charlie'])

# üìô DataFrame 2: Student names as index, and Science marks
df_science = pd.DataFrame({
    'Science_Marks': [88, 81, 95, 67]
}, index=['Alice', 'Bob', 'David', 'Kevin'])

# üîó Using join (index-based join)
joined_df = df_math.join(df_science)
print("üîπ Default join (LEFT JOIN on index):")
print(joined_df)

# --- If we did merge instead ---
# Since df_math and df_science don't have 'Name' as a column but as index,
# merge would not automatically work without resetting the index or specifying left_index=True, right_index=True.
# For example:
# merged_df = pd.merge(df_math, df_science, left_index=True, right_index=True, how='left')
# The output would be the same as join in this case:
#       Math_Marks  Science_Marks
# Alice          85            88
# Bob            78            81
# Charlie        92           NaN

# üîó INNER JOIN with join
joined_inner = df_math.join(df_science, how='inner')
print("\nüîπ INNER JOIN (on index):")
print(joined_inner)

# --- Using merge with inner join on indexes ---
# merged_inner = pd.merge(df_math, df_science, left_index=True, right_index=True, how='inner')
# Output would be:
#       Math_Marks  Science_Marks
# Alice          85            88
# Bob            78            81

# üîó OUTER JOIN with join
joined_outer = df_math.join(df_science, how='outer')
print("\nüîπ OUTER JOIN (on index):")
print(joined_outer)

# --- Using merge with outer join on indexes ---
# merged_outer = pd.merge(df_math, df_science, left_index=True, right_index=True, how='outer')
# Output would be:
#       Math_Marks  Science_Marks
# Alice          85          88.0
# Bob            78          81.0
# Charlie        92           NaN
# David         NaN          95.0

# üîó RIGHT JOIN with join
joined_right = df_math.join(df_science, how='right')
print("\nüîπ RIGHT JOIN (on index):")
print(joined_right)

# --- Using merge with right join on indexes ---
# merged_right = pd.merge(df_math, df_science, left_index=True, right_index=True, how='right')
# Output would be:
#       Math_Marks  Science_Marks
# Alice          85          88.0
# Bob            78          81.0
# David         NaN          95.0
```

What is concat?

concat means concatenate or stick together.

It‚Äôs used to combine multiple DataFrames or Series by stacking them either vertically (one below another) or horizontally (side by side).

It does not merge based on keys or indexes like merge or join.

Instead, it just joins data along a particular axis (rows or columns).

Why is concat useful?

When you have multiple pieces of similar data split into different DataFrames (e.g., monthly sales data) and want to combine them into one big DataFrame.

When you want to add new columns or rows quickly without worrying about matching keys.

Useful for quick stacking or appending operations.



```python
import pandas as pd

# üéì DataFrame A: First batch of students with their Math marks
df_a = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92]
})

# üéì DataFrame B: Second batch of students with their Math marks
df_b = pd.DataFrame({
    'Name': ['David', 'Eva'],
    'Math': [88, 79]
})

# üîΩ 1. Vertical Concatenation (axis=0): Stack df_b under df_a
df_all_math = pd.concat([df_a, df_b])

# üìù This combines both batches into one big DataFrame.
# Index is NOT reset automatically ‚Äî still uses original row numbers.
print("üîπ All Math Scores (Vertical Concat):")
print(df_all_math)


# üìò DataFrame C: Science marks for the same 5 students
df_science = pd.DataFrame({
    'Science': [91, 84, 89, 76, 85]
}, index=[0, 1, 2, 3, 4])  # Indices match the default order of df_all_math

# üîÅ 2. Horizontal Concatenation (axis=1): Add Science marks beside Math
df_all_marks = pd.concat([df_all_math.reset_index(drop=True), df_science], axis=1)

# üìù We reset index in df_all_math so it aligns perfectly with df_science.
print("\nüîπ Math + Science Scores (Horizontal Concat):")
print(df_all_marks)
```

When to use Vertical concat(axis=0):

When column names are the same.

You want to add more rows (like adding more students to a list).

Think: "Same structure, more data."

 When to use Horizontal concat(axis=1):
 
When column names are different (e.g., "Math", "Science").

You want to add more information/subjects to the same rows.

Think: "Same people, more details."

 What is append()?

append() is a method used to add rows of one DataFrame to another.

It's like vertical stacking ‚Äî similar to concat(axis=0).

üîπ What it does:

Adds the rows of another DataFrame or Series to the end of the original DataFrame.

Returns a new DataFrame; original one stays unchanged unless reassigned.

It matches columns by name ‚Äî if a column is missing in either, it fills with NaN.

üîπ Types:

Technically, append() has one use case ‚Äî adding rows.

But it works in two ways:

Appending a DataFrame ‚Üí adds all rows of another DataFrame.

Appending a Series or dictionary ‚Üí adds a single row.

‚ö†Ô∏è Note:

append() is deprecated since pandas 1.4.0 and removed in 2.0.

Use pd.concat([...]) instead ‚Äî it is more flexible and recommended.

```python
import pandas as pd

# üßë First DataFrame: Student names with Math scores
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Math': [85, 78]
})

# üë®‚Äçüéì Second DataFrame: More students with Math scores
df2 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Math': [92, 88]
})

# ‚úÖ RECOMMENDED: Use pd.concat() to append df2 below df1
# axis=0 means rows are being added (vertical stacking)
df_combined = pd.concat([df1, df2], ignore_index=True)

# üìù ignore_index=True: Resets the index in the new DataFrame (0, 1, 2, ...)
print("üîπ Combined DataFrame using concat():")
print(df_combined)
# ‚ùå NOT RECOMMENDED: Using append() (deprecated in future versions)
# df_combined_append = df1.append(df2, ignore_index=True)
```
